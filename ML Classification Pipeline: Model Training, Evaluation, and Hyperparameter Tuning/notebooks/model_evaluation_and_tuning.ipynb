```python
from sklearn.linear_model import LogisticRegression

log_model = LogisticRegression(max_iter=1000, random_state=42)

# Train the Model
log_model.fit(X_train, y_train)
```

# Making sure that the data is scaled

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

# During Logistic Regression training, the optimizer reached the maximum number of iterations before convergence.
# The maximum number of iterations were increased to 5000, allowing the optimization algorithm sufficient iterations 
# to converge and successfully train the model.

```python
# INCREASING THE LIMIT

log_model = LogisticRegression(
    max_iter=5000,
    random_state=42
)
```

```python
# Train the model
log_model.fit(X_train, y_train)
```

```python
# Predict on the test data
y_pred_log = log_model.predict(X_test)
```

```python
# Calculate accuracy
log_accuracy = accuracy_score(y_test, y_pred_log)

print(f"Logistic Regression Accuracy: {log_accuracy:.2f}")
```

# Evaluating both baseline models using key metrics: Accuracy, Precision, Recall, and F1-Score for classification

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import pandas as pd
```

# (a) Comparing their performances

```python
comparison = pd.DataFrame({
    "Metric": ["Accuracy", "Precision", "Recall", "F1-Score"],
    "Logistic Regression": [
        accuracy_score(y_test, y_pred_log),
        precision_score(y_test, y_pred_log, average='weighted'),
        recall_score(y_test, y_pred_log, average='weighted'),
        f1_score(y_test, y_pred_log, average='weighted')
    ],
    "Decision Tree": [
        accuracy_score(y_test, y_pred_tree),
        precision_score(y_test, y_pred_tree, average='weighted'),
        recall_score(y_test, y_pred_tree, average='weighted'),
        f1_score(y_test, y_pred_tree, average='weighted')
    ]
})

comparison
```

```python
import matplotlib.pyplot as plt

comparison.set_index("Metric").plot(
    kind="bar",
    figsize=(8, 5)
)

plt.title("Comparison of Logistic Regression and Decision Tree")
plt.ylabel("Score")
plt.ylim(0, 1)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

# Performing Hyperparameter Tuning on Logistic Regression Model using GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
```

```python
# Defining the Hyperparameter Grid

param_grid = {
    'C': [0.01, 0.1, 1, 10, 100],
    'solver': ['lbfgs', 'liblinear'],
    'penalty': ['l2'],
    'max_iter': [1000, 3000, 5000]
}
```

```python
# Creating GridSearchCV

grid_search = GridSearchCV(
    estimator=LogisticRegression(random_state=42),
    param_grid=param_grid,
    cv=5,
    scoring='accuracy',
    n_jobs=-1,
    verbose=1
)
```

```python
# Training GridSearchCV
grid_search.fit(X_train, y_train)
```

```python
print("Best Parameters:")
print(grid_search.best_params_)

print("\nBest Cross-Validation Accuracy:")
print(grid_search.best_score_)
```

```python
# Evaluating Best Model
best_model = grid_search.best_estimator_

y_pred_best = best_model.predict(X_test)

print("BEST MODEL")
print(best_model)
```

```python
# Calculating Accuracy

from sklearn.metrics import accuracy_score

best_accuracy = accuracy_score(y_test, y_pred_best)

print(f"Tuned Logistic Regression Accuracy: {best_accuracy:.2f}")
```

# Comparison of Accuracy of Model before and after Tuning

```python
import pandas as pd

comparison = pd.DataFrame({
    "Model": ["Baseline Logistic Regression",
              "Tuned Logistic Regression"],
    "Accuracy": [
        log_accuracy,
        best_accuracy
    ]
})

print(comparison)
```
